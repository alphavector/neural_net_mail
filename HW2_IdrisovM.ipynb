### 2. Модели

```
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score
```

```
#mnist = fetch_mldata('MNIST original')
!wget -O mnist.dump https://www.dropbox.com/s/756ltmrktbqtd4r/mnist.dump?dl=0
import pickle
with open("mnist.dump","rb") as f:
  mnist = pickle.load(f, encoding='latin1')
```

```
X_mnist = mnist.data
y_mnist = mnist.target

X_train, X_test, y_train, y_test = train_test_split(X_mnist, 
                                                                            y_mnist, 
                                                                            random_state = 42,
                                                                            stratify = y_mnist)
```

```
X_train_normalised = X_train/255.0
X_test_normalised = X_test/255.0
```

```
%%time
clf_svc = SVC(C=200,kernel='rbf',gamma=0.01,cache_size=8000,probability=False)
clf_svc.fit(X_train_normalised,  y_train)
print("Точность на обучающей выборке: {}".format(clf_svc.score(X_train_normalised,  y_train)))
print("Точность на тестовой выборке: {}".format(clf_svc.score(X_test_normalised,  y_test)))
```

```
%%time
clf_rfc = RandomForestClassifier(random_state=47)
clf_rfc.fit(X_train_normalised,  y_train)
print("Точность на обучающей выборке: {}".format(clf_rfc.score(X_train_normalised,  y_train)))
print("Точность на тестовой выборке: {}".format(clf_rfc.score(X_test_normalised,  y_test)))
```

```
%%time
clf_abc = AdaBoostClassifier()
clf_abc.fit(X_train_normalised,  y_train)
print("Точность на обучающей выборке: {}".format(clf_abc.score(X_train_normalised,  y_train)))
print("Точность на тестовой выборке: {}".format(clf_abc.score(X_test_normalised,  y_test)))
```

```
%%time
clf_dtc = DecisionTreeClassifier()
clf_dtc.fit(X_train_normalised,  y_train)
print("Точность на обучающей выборке: {}".format(clf_dtc.score(X_train_normalised,  y_train)))
print("Точность на тестовой выборке: {}".format(clf_dtc.score(X_test_normalised,  y_test)))
```

```
%%time
param_list = [("eta", 0.08), ("max_depth", 6), ("subsample", 0.8), ("colsample_bytree", 0.8), ("objective", "multi:softmax"), ("eval_metric", "merror"), ("alpha", 8), ("lambda", 2), ("num_class", 10)]
n_rounds = 600
early_stopping = 50
    
d_train = xgb.DMatrix(X_train_normalised, label=y_train)
d_val = xgb.DMatrix(X_test_normalised, label=y_test)
eval_list = [(d_train, "train"), (d_val, "test")]
bst = xgb.train(param_list, d_train, n_rounds, evals=eval_list, early_stopping_rounds=early_stopping, verbose_eval=True)
```

```
y_pred_train = bst.predict(d_train)
y_pred_test = bst.predict(d_val)
```

```
print("Точность на обучающей выборке: ", accuracy_score(y_pred_train, y_train))
print("Точность на тестовой выборке: ", accuracy_score(y_pred_test, y_test))
```
